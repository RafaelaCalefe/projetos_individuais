## EDA : Análise bivariada + Teste de Hipótese


1. Perfil de idade das pessoas com sintomas de depressão

```python
sns.boxplot(x= "phq_grp", y= "Idade", data=combinacao_2 ) #Boxplot da categoria de sintomas de depressão com Idade
```

Para o Teste de Hipótese iremos comparar a média de idade entre os grupos: 1, 2, 3, 4 e 5. Usaremos Teste F da ANOVA. H0:mu0 = mu1 = mu2 = mu3 = mu4 vs H1: pelo menos 1 diferença

```python
from scipy.stats import f_oneway
```

```python
combinacao_aux = combinacao_2[["phq_grp", "Idade"]].dropna()
```

```python
stat, p = f_oneway(combinacao_aux[(combinacao_aux.phq_grp == 0)]["Idade"],
                   combinacao_aux[(combinacao_aux.phq_grp == 1)]["Idade"],
                   combinacao_aux[(combinacao_aux.phq_grp == 2)]["Idade"],
                   combinacao_aux[(combinacao_aux.phq_grp == 3)]["Idade"],
                   combinacao_aux[(combinacao_aux.phq_grp == 4)]["Idade"])

print("stat=%.3f, p=%.3f" % (stat,p))
                  
                  
```

Aceitamos H0 pois p=0.402 que é maior que 5%

2. Perfil de renda das pessoas com sintomas de depressão

```python
sns.boxplot(x= "phq_grp", y= "Renda anual fam", data=combinacao_2 ) #Boxplot da categoria de sintomas de depressão com renda
```

```python
combinacao_aux = combinacao_2[["phq_grp", "Renda anual fam"]].dropna()
```

```python
stat, p = f_oneway(combinacao_aux[(combinacao_aux.phq_grp == 0)]["Renda anual fam"],
                   combinacao_aux[(combinacao_aux.phq_grp == 1)]["Renda anual fam"],
                   combinacao_aux[(combinacao_aux.phq_grp == 2)]["Renda anual fam"],
                   combinacao_aux[(combinacao_aux.phq_grp == 3)]["Renda anual fam"],
                   combinacao_aux[(combinacao_aux.phq_grp == 4)]["Renda anual fam"])

print("stat=%.3f, p=%.5f" % (stat,p))
                  
```

Rejeitamos H0 pois p = 0 que é menor que 5%.
Como rejeitamos, devemos agora investigar onde está acontecendo essa diferença, para isso utilizaremos o teste de Tuckey

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(combinacao_aux["Renda anual fam"],
    combinacao_aux["phq_grp"],
    alpha = 0.5)
print(tukey)
```
